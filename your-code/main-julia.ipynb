## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
from scipy.stats import uniform

# Define the range
a = 0  # bottom
b = 0  # ceiling
count = 0  # count

# Generate uniform random numbers between a and b
random_numbers = uniform.rvs(loc=a, scale=b - a, size=count)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
# Function to generate uniformly distributed random numbers
def generate_uniform_random_numbers(bottom, ceiling, count):
    # Generate uniform random numbers between 'bottom' and 'ceiling'
    random_numbers = uniform.rvs(loc=bottom, scale=ceiling - bottom, size=count)
    return random_numbers

# Call the function with the first set of parameters
data1 = generate_uniform_random_numbers(bottom=10, ceiling=15, count=100)

# Call the function with the second set of parameters
data2 = generate_uniform_random_numbers(bottom=10, ceiling=60, count=1000)

# Plot the histograms
plt.figure(figsize=(12, 6))

# First histogram
plt.subplot(1, 2, 1)
plt.hist(data1, bins=10, alpha=0.7)
plt.title('Uniform Distribution (100 samples)')
plt.xlabel('Value')
plt.ylabel('Count')
plt.ylim(0, 120) 

# Second histogram
plt.subplot(1, 2, 2)
plt.hist(data2, bins=10, alpha=0.7)
plt.title('Uniform Distribution (1000 samples)')
plt.xlabel('Value')
plt.ylabel('Count')
plt.ylim(0, 120) 

# Display the plots
plt.tight_layout()
plt.show()
```

How are the two distributions different?

```python
# Both distributions are uniform but the left one is spread across a narrow x-axis and has bars with smaller hights, 
# whereas the right one has bigger bins and accommodates a larger range of values.
# You could say the left histogram looks morge bumpy due to the limited range of values whereas the right one looks more uniform.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to generate normally distributed random numbers
def generate_normal_random_numbers(mean, std_dev, count):
    return np.random.normal(loc=mean, scale=std_dev, size=count)

# Generate 1,000 normally distributed numbers with mean=10, std_dev=1
data3 = generate_normal_random_numbers(mean=10, std_dev=1, count=1000)

# Generate 1,000 normally distributed numbers with mean=10, std_dev=50
data4 = generate_normal_random_numbers(mean=10, std_dev=50, count=1000)

# Plot the histograms of the distributions
plt.figure(figsize=(12, 6))

# Plot for the first distribution (std_dev=1)
plt.subplot(1, 2, 1)
plt.hist(data3, bins=30, alpha=0.7)
plt.title('Normal Distribution (std dev = 1)')
plt.xlabel('Value')
plt.ylabel('Count')

# Plot for the second distribution (std_dev=50)
plt.subplot(1, 2, 2)
plt.hist(data4, bins=30, alpha=0.7)
plt.title('Normal Distribution (std dev = 50)')
plt.xlabel('Value')
plt.ylabel('Count')

# Display the plots
plt.tight_layout()
plt.show()
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# Function to generate exponentially distributed numbers
def generate_exponential_numbers(mean, size):
    return np.random.exponential(scale=mean, size=size)

# Generate two sequences
sequence1 = generate_exponential_numbers(mean=10, size=10)
sequence2 = generate_exponential_numbers(mean=10, size=100)

# Plotting the histograms
plt.figure(figsize=(12, 6))

# Histogram for sequence of size 10
plt.subplot(1, 2, 1)
plt.hist(sequence1, bins=100)
plt.title('Exponential Distribution (size=10)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Histogram for sequence of size 100
plt.subplot(1, 2, 2)
plt.hist(sequence2, bins=100) 
plt.title('Exponential Distribution (size=100)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

How are the two distributions different?

The mean changes, so the distribution changes as well. 

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# Cumulative distribution function: P(X > x) = 1 - P(X \leq x)
```

```python
from scipy.stats import expon

mean_time = 10  # Mean time spent in the bank (in minutes)
rate_lambda = 1 / mean_time  # Rate parameter λ
time_limit = 15  # Time in minutes for which we want to find the probability

# Calculate the CDF for the given time limit
probability = expon.cdf(time_limit, scale=1/rate_lambda)

# Display the result
print(f"The probability that a customer spends less than {time_limit} minutes in the bank is: {probability:.4f}")
```

What is the probability that the customer will spend more than 15 minutes?

```python
mean_time = 10  # Mean time spent in the bank (in minutes)
rate_lambda = 1 / mean_time  # Rate parameter λ
time_limit = 15  # Time in minutes for which we want to find the probability

# Calculate the CDF for the given time limit
probability_less_than_15 = expon.cdf(time_limit, scale=1/rate_lambda)

# Calculate the probability of spending more than 15 minutes
probability_more_than_15 = 1 - probability_less_than_15

# Display the result
print(f"The probability that a customer spends more than {time_limit} minutes in the bank is: {probability_more_than_15:.4f}")
```
